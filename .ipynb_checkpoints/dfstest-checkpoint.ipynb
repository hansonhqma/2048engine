```python
import networkx as nx
import numpy as np
from game import engine
```

```python
import matplotlib.pyplot as plt
```

```python
exp = nx.Graph()
env = engine()
```

```python
actions = ('n', 's', 'e', 'w')
size = 4
node_labels = dict()
edge_labels = dict()
```

```python
def search(game:engine, depth:int, root=True):
    if root:
        exp.clear()
        node_labels.clear()
        edge_labels.clear()
    
    if depth == 0:
        # base case
        node_labels[game] = game.score()
        return game.score()
    
    values = np.zeros((size))
    
    for i, action in enumerate(actions):
        copy = game.replicate()
        valid_move = copy.move(action)
        if not valid_move:
            continue
            
        
        v = search(copy, depth-1, root=False)
        exp.add_edge(game, copy)
        edge_labels[(game, copy)] = action
        values[i] = v
    
    node_labels[game] = int(np.max(values))
    
    if root:
        return np.argmax(values), np.max(values)
    else:
        return np.max(values)
```

```python
search(env, 2)
```

```python
plt.figure(dpi=800)
pos = nx.spring_layout(exp)
nx.draw_networkx_labels(exp, pos, font_size=3)
nx.draw_networkx_edge_labels(exp, pos, edge_labels, font_size=5, font_color='red')
nx.draw(exp, pos, node_color='yellow')
```
